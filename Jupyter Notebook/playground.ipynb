```python
def quickSort(arr, s, e ):
    if s>=e:
        return None
    v = arr[(s+e)//2]
    pivot = findPivot(arr, s, e, v)
    quickSort(arr, s, pivot-1)
    quickSort(arr,pivot, e)
    
    
def findPivot(arr, s, e, v):
    l = s
    r = e 
    while l<=r:
        while arr[l] < v:
            l+=1
        while arr[r] > v:
            r-=1
        if l<=r:
            swap(arr, l, r)
            l+=1
            r-=1      
    return l
    
def swap(arr, a, b):
    t = arr[a]
    arr[a] = arr[b]
    arr[b]= t
    
arr = [1,3,1,6,2,76,2,3]
quickSort(arr, 0, len(arr)-1)
print(arr) 
```

```python
def heapify(arr, s, e):
    c1 = 2*s +1
    c2 = c1+1
    maxx = s
    if c1 <= e and arr[c1]>arr[maxx]:
        maxx = c1
    if c2<=e and arr[c2]> arr[maxx]:
        maxx = c2
    if maxx!=s:
        swap(arr, maxx, s)
        heapify(arr, maxx, e)
        
def build_heap(arr):
    s = (len(arr)-1)//2
    e = len(arr)-1
    while s>-1:
        heapify(arr, s, e) 
        s-=1
def heapsort(arr):
    build_heap(arr)
    i = len(arr)-1
    while i > 0:
        swap(arr, 0, i)
        heapify(arr, 0, i-1)
        i-=1
        
arr = [1,3,1,6,2,76,2,3]
heapsort(arr)
print(arr)     
    
    
    
    
    
    
    
    
    
    
    
    
```
